```python
import random
```

```python
N = 90000 
gc = 0.6
motif = "ATAGCCGA"
```

```python
gc_len = round(len(motif) * gc)
at_len = len(motif) - gc_len
```

```python
motifs = []
for i in range(N):
    letters = []
    for _ in range(gc_len):
        letters.append(random.choice(("G", "C")))
    for _ in range(at_len):
        letters.append(random.choice(("A", "T")))
    motifs.append(''.join(random.sample(letters, len(letters))))        
```

```python
motifs[0:3]
```

```python
checks = 0
for m in motifs:
    print(m)
    if m == motif:
        checks += 1
        
checks / N
```
